#Task 1: Conceptual Understanding

1.What is the difference between "Love" and "love" in NLP?
>In NLP, “Love” and “love” are treated as different tokens due to case sensitivity. This can lead to inconsistent analysis. Converting text to lowercase ensures uniformity and prevents duplication of words with the same meaning.

2.What happens if stopwords are not removed?
>If stopwords are not removed, the dataset becomes noisy and contains many irrelevant words like “is”, “the”, and “and”. This reduces model efficiency, increases computational cost, and may negatively affect performance by diluting meaningful information.

3.Provide two real-world scenarios where removing stopwords can be harmful.
>Removing stopwords can harm sentiment analysis, where words like “not” change meaning (e.g., “not happy”). It is also harmful in question answering systems, where words like “what”, “when”, and “how” are essential for understanding context.

4.Explain the difference between stemming and lemmatization.
>Stemming reduces words to their root form by removing suffixes, often producing non-meaningful words. Lemmatization uses vocabulary and linguistic rules to return the correct base form, making it more accurate and suitable for real-world NLP applications.

```python

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""
  # Tokenization
    tokens = text.split()

    # Remove short words (<=2), keep "no", "not"
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)

    # Join back to sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence

sentence = "I am not happy with this!!"
tokens, cleaned = preprocess_text(sentence)

print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned)
```

```python

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    return text
print(preprocess_text("Contact me at test@gmail.com"))
print(preprocess_text("Visit https://innomatics.com now"))
```

#Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


for sentence in test_sentences:
    processed_text_output = preprocess_text(sentence)
    tokens = processed_text_output.split()
    cleaned = processed_text_output

    print("Original Text: ", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 60)
```

#Task 4: Token Analytics


```python
def token_analysis(tokens):

    if len(tokens) == 0:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


for sentence in test_sentences:
    processed_text_output = preprocess_text(sentence)
    #tokenize this string to get the 'tokens' list
    tokens = processed_text_output.split()

    total, unique, avg_len = token_analysis(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
    print("-" * 60)
```

#Task 6: Build Full Pipeline

```python

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        processed_text_output = preprocess_text(text)
        tokens = processed_text_output.split()
        cleaned = processed_text_output
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
